```python
#WEB SCRAPING ASSIGNMENT-1

#QUESTION 1
#Write a python program to display all the header tags from wikipedia.org.

!pip install bs4
!pip install requests

from bs4 import BeautifulSoup
import requests

page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
page

soup = BeautifulSoup(page.content)
soup

head = []
for i in soup.find_all('span',class_="mw-headline"):
    head.append(i.text)

head

import pandas as pd
df = pd.DataFrame({'HEADER TAGS':head})
df
```

```python
#QUESTION 2
#Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

!pip install bs4
!pip install requests

from bs4 import BeautifulSoup
import requests

page = requests.get('https://www.imdb.com/list/ls091520106/')
page

soup = BeautifulSoup(page.content)
soup

movie = []
for i in soup.find_all('h3',class_="lister-item-header",):
    movie.append(i.text.replace('\n',''))

movie

year = []
for i in soup.find_all('span',class_="lister-item-year text-muted unbold",):
    year.append(i.text)
year 

rating = []
for i in soup.find_all('div',class_="ipl-rating-star small"):
    rating.append(i.text.replace('\n',''))
rating

print(len(movie),len(rating),len(year))

import pandas as pd
df = pd.DataFrame({'movie':movie,'rating':rating,'Year':year})
df

```

```python
#QUESTION 4
#Write a python program to scrape product name, price and discounts from https://meesho.com/bags-ladies/pl/p7vbp .

!pip install bs4
!pip install requests

from bs4 import BeautifulSoup
import requests

page = requests.get('https://meesho.com/bags-ladies/pl/p7vbp')
page

soup = BeautifulSoup(page.content)
soup

name = []
for i in soup.find_all('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
    name.append(i.text)
    
name

price = []
for i in soup.find_all('h5',class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
    price.append(i.text.replace('₹',''))
price

dis = []
for i in soup.find_all('span',class_="Text__StyledText-sc-oo0kvp-0 lnonyH"):
    dis.append(i.text.replace('off',''))
dis

print(len(name),len(dis),len(price))

import pandas as pd
df = pd.DataFrame({'Product name':name,'price':price,'Discount':dis})
df
```

```python
#QUESTION 7
#Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
#and the code for the video from the link for the youtube video from the post.

!pip install bs4
!pip install requests

from bs4 import BeautifulSoup
import requests

page = requests.get('https://coreyms.com/')
page

soup = BeautifulSoup(page.content)
soup

header = []
for i in soup.find_all('h2',class_="entry-title"):
    header.append(i.text)
    
header

date = []
for i in soup.find_all('time',class_="entry-time"):
    date.append(i.text)
date

con = []
for i in soup.find_all('div',class_="entry-content"):
    con.append(i.text.replace('\n',''))
con

print(len(header),len(date),len(con))
import pandas as pd
df = pd.DataFrame({'Heading':header,'date':date,'content':con})
df

```

```python
#QUESTION 9
#Write a python program to scrape mentioned details from dineout.co.in
#(1)restaurant name (2)cuisine (3)location (4)ratings (5)image url
!pip install bs4
!pip install requests

from bs4 import BeautifulSoup
import requests

page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page

soup = BeautifulSoup(page.content)
soup

Restaurant_name = []
for i in soup.find_all('div',class_="restnt-info cursor"):
    Restaurant_name.append(i.text)
    
Restaurant_name

location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location

rating = []
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
rating

cuisine = []
for i in soup.find_all('div',class_="detail-info"):
    cuisine.append(i.text.split()[6])
cuisine

imageURL = []
for i in soup.find_all('img',class_="no-img"):
    imageURL.append(i['data-src'])
imageURL

print(len(Restaurant_name),len(cuisine),len(location),len(rating),len(imageURL))

import pandas as pd
df = pd.DataFrame({'Restaurant':Restaurant_name,'cuisine':cuisine,'location':location,'ratings':rating,'imageURL':imageURL})
df
```
